```python
import numpy as np
```

```python
input = np.array([[-2,0],[0,2],[-2,2]])
output = np.array([1,1,0])
```

```python
class Perceptron:
    def __init__(self, input, output):
        self.input = input
        self.output = output
        self.weights = np.random.rand(input.shape[1],1)
        self.bias = np.random.rand(1)
        self.learning_rate = 0.1
        self.epochs = 100
        self.error = []
    def activation(self, x):
        if x > 0:
            return 1
        else:
            return 0
    def train(self):
        for i in range(self.epochs):
            for j in range(self.input.shape[0]):
                y = np.dot(self.input[j], self.weights) + self.bias
                y = self.activation(y)
                err = self.output[j] - y
                self.weights = self.weights+ err * np.transpose([self.input[j]])
                self.bias = self.bias + err
                if err != 0:
                    print('false')
                else:
                    print('true')
```

```python
perceptron = Perceptron(input, output)
```

```python
perceptron.train()
```
